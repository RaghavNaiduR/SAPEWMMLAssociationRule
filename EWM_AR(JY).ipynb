```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
import random
```

```python
store_data = pd.read_excel (r'C:\Bizadept\ML\1.xlsx', header=None)
store_data.head()
print(store_data)
records = []
random.shuffle(records)
for i in range(1, 59):
     records.append([str(store_data.values[i,j]) for j in range(1, 5)])
for i,j in enumerate(records):
    while 'nan' in records[i]: records[i].remove('nan')
print(records)
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.5, min_lift=4, min_length=2)
association_rules = list(association_rules)
print(len(association_rules))
```
